```python
import pandas as pd
import pygraphviz as pgv
```

```python
schema = pd.read_csv('schema.csv', sep='\t', header=None)
```

```python
df = pd.read_csv('http://data.gdeltproject.org/gdeltv2/20150218224500.translation.gkg.csv.zip', compression='zip', header=None, sep='\t', quotechar='"')
```

```python
schema[0]
```

```python
df.columns = list(schema[0].values)
```

```python
df.head(10)
```

```python
sample_df = df
```

```python
people_dict = {}
G = pgv.AGraph()
```

```python
for ix, data in sample_df.iterrows():
    url = str(data['DocumentIdentifier'])
    people = str(data['Persons']).split(';')

    # update url list
    for person in people:
        if person == 'nan':
            continue
        if person in people_dict:
            people_dict[person]['urls'].append(url)
        else:
            people_dict[person] = {
                'name': person,
                'connections': {},
                'urls': [url],
                'sentiment': 0,
            }
    
    # update connection list
    for p1 in people:
        for p2 in people:
            if p1 == 'nan' or p2 == 'nan':
                continue
            if p1 == p2:
                continue
            if p2 in people_dict[p1]['connections']:
                people_dict[p1]['connections'][p2]['urls'].append(url)
                people_dict[p1]['connections'][p2]['strength'] += 1
            else:
                people_dict[p1]['connections'][p2] = {
                    'urls': [url],
                    'strength': 1
                }
```

```python
def to_edge_list(start_node, connections):
    '''
    Converts connections dict to edge list for networkx
    '''
    edges = []
    for end_node in connections.keys():
        edges.append((start_node, end_node))
    return edges
```

```python
for p in people_dict.keys():
    G.add_node(p)
    G.add_edges_from(to_edge_list(p, people_dict[p]['connections']))
```

```python
import matplotlib.pyplot as plt
```

```python
G.layout()
G.draw('network.svg')
```
